## 2D Line Plot

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 


hours = [ 2,4,6,4,5,7,4]
days = [1,2,3,4,5,6,7]
plt.plot(days,hours)
```

```python
plt.xlabel('Days')
plt.ylabel('Study Hours') 
plt.title('A student study Time over a week')
plt.plot(days,hours)
```

```python
# multiple student information 
hours_1 = [2,4,3,6,1,5,0]
hours_2 = [3,4,5,4,1,4,2]
days = [1,2,3,4,5,6,7]
plt.xlabel('Days')
plt.ylabel('Study Hours') 
plt.title('Students study Time over a week')
plt.plot(days,hours_1)
plt.plot(days,hours_2)
```

## 2D Line from Pandas DataFrame

```python
df = pd.read_csv('enrollment_data.csv')
df
```

```python
plt.plot(df['Year'],df['Programming'])
```

```python
plt.xlabel('Years')
plt.ylabel('Total Enrollment') 
plt.title('Students Learning Programming Last 8 years')
plt.plot(df['Year'],df['Programming'])
```

```python
plt.xlabel('Years')
plt.ylabel('Total Enrollment') 
plt.title('Students Learning Digital Marketing Last 8 years')
plt.plot(df['Year'],df['Digital Marketing'])
```

```python
plt.plot(df['Year'],df['Programming'])
plt.plot(df['Year'],df['Digital Marketing'])
```

```python
plt.xlabel('Years')
plt.ylabel('Total Enrollment') 
plt.title('Students Learning Programming and  Digital Marketing Last 8 years')
plt.plot(df['Year'],df['Programming'])
plt.plot(df['Year'],df['Digital Marketing'])
```

```python
# adding labels usinf plt.legend()

plt.xlabel('Years')
plt.ylabel('Total Enrollment') 
plt.title('Students Learning Programming and  Digital Marketing Last 8 years')
plt.plot(df['Year'],df['Programming'],label = 'Programming')
plt.plot(df['Year'],df['Digital Marketing'],label = 'Digital Marketing')
plt.legend()
```

```python
# adding labels usinf plt.legend()
# positioning the label 

plt.xlabel('Years')
plt.ylabel('Total Enrollment') 
plt.title('Students Learning Programming and  Digital Marketing Last 8 years')
plt.plot(df['Year'],df['Programming'],label = 'Programming')
plt.plot(df['Year'],df['Digital Marketing'],label = 'Digital Marketing')
plt.legend(loc = 'lower right')
```

```python
# adding labels usinf plt.legend()
# positioning the label 

plt.xlabel('Years')
plt.ylabel('Total Enrollment') 
plt.title('Students Learning Programming and  Digital Marketing Last 8 years')
plt.plot(df['Year'],df['Programming'],label = 'Programming')
plt.plot(df['Year'],df['Digital Marketing'],label = 'Digital Marketing')
plt.legend(loc = 'center right')
```

```python
# adding labels usinf plt.legend()
# positioning the label 

plt.xlabel('Years')
plt.ylabel('Total Enrollment') 
plt.title('Students Learning Programming and  Digital Marketing Last 8 years')
plt.plot(df['Year'],df['Programming'],label = 'Programming')
plt.plot(df['Year'],df['Digital Marketing'],label = 'Digital Marketing')
plt.legend(loc = 'best')
```
